```python
# Import the necessary libraries
## Extracing and processing country indicators from https://data.un.org/en/index.html
import requests #for extracting web content
import pandas as pd, numpy as np #for data manipulation
from bs4 import BeautifulSoup #for parsing web content
```

```python
#import country names data
country_names = pd.read_csv("Countries1.csv")

#extract alphas column, converting values to lower case
country_codes = country_names['Alpha2'].str.lower()
```

```python
#1) Extract tabular page content
def extract_tabular_content(url):
    html_content = requests.get(url)

    # create BeautifulSoup object
    soup = BeautifulSoup(html_content.text, "html.parser")

    # extract the neccessary content parts (here it is in the <uL> element) and the tables with class name 'pure-table'
    ul_content = soup.find("ul")
    tbl_content = ul_content.find_all("table", class_="pure-table")
    return tbl_content
```

```python
#Build the URL structure:  https://data.un.org/en/iso/af.html
url_stab = "https://data.un.org/en/iso/"
country_alpha = ""
url_suffix = ".html"
af_url = url_stab + country_codes[0] + url_suffix
```
